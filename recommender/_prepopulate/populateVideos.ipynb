```python
from dotenv import load_dotenv
import os
import requests

load_dotenv()
api_key = os.environ['TMDB_API_KEY']
```

```python
def trailers(tmdbId):
    response = requests.get(f'https://api.themoviedb.org/3/movie/{tmdbId}/videos?api_key={api_key}')
    return [{'name': result['name'], 'key':result['key']} for result in response.json()['results'] if result['type'] == 'Trailer' and result['site'] == 'YouTube']

trailers(15602)
```

```python
import pandas as pd

df = pd.read_csv('../data/full/links_edited.csv')
df
```

```python
for i, row in df.iterrows():
    try:
        movieId = row['movie_idx']
        tmdbId = row['tmdbId']
        print(i, end='\r')
        requests.put(f'http://localhost:3001/movies/{movieId}', json={'trailers': trailers(tmdbId)})
    except Exception as e:
        print(repr(e), row['movie_idx'])
```
